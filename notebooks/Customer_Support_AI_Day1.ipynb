```python
import pandas as pd

# Load dataset
df = pd.read_csv('/content/customer_support_tickets.csv')

# Show first 5 rows
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Drop unnecessary columns
df = df.drop(['Customer Name', 'Customer Email', 'Ticket ID'], axis=1)

df.head()
```

```python
# Drop rows where important columns are missing
df = df.dropna(subset=['Ticket Type', 'Ticket Priority', 'Ticket Description'])

df.shape
```

```python
df_clean = df.copy()
```

```python
# Convert to datetime
df_clean['Date of Purchase'] = pd.to_datetime(df_clean['Date of Purchase'])

# Check
df_clean.info()
```

```python
# Extract date features
df_clean['purchase_year'] = df_clean['Date of Purchase'].dt.year
df_clean['purchase_month'] = df_clean['Date of Purchase'].dt.month
df_clean['purchase_day'] = df_clean['Date of Purchase'].dt.day

df_clean.head()
```

```python
# Text length features
df_clean['desc_length'] = df_clean['Ticket Description'].apply(len)
df_clean['subject_length'] = df_clean['Ticket Subject'].apply(len)

df_clean.head()
```

```python
df_clean = df_clean.drop('Date of Purchase', axis=1)
```

```python
# Save cleaned dataset
df_clean.to_csv('cleaned_data_day1.csv', index=False)
```
