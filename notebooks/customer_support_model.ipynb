```python
# STEP 1: Import libraries
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import pickle

# STEP 2: Data
X = [
    # Refund
    "refund my order",
    "order refund request",
    "refund for my order",
    "please process my refund",
    "cancel order and refund",
    "refund not received yet",
    "i want refund for order",

    # Billing
    "billing issue",
    "incorrect bill amount",
    "charged extra",
    "payment problem",

    # Technical
    "app not opening",
    "error while using app",
    "app crashes",
    "technical issue",

    # Delivery
    "order not delivered",
    "delivery is late",
    "where is my order",
    "order delayed",

    # Product
    "product details",
    "tell me about product",
    "product information",
    "explain this product"
]

y = [
    "refund","refund","refund","refund","refund","refund","refund",
    "billing","billing","billing","billing",
    "technical","technical","technical","technical",
    "delivery","delivery","delivery","delivery",
    "product","product","product","product"
]

# STEP 3: Vectorizer
vectorizer = TfidfVectorizer()
X_vec = vectorizer.fit_transform(X)

# STEP 4: Model
model = LogisticRegression()
model.fit(X_vec, y)

# STEP 5: Save files
pickle.dump(model, open("model.pkl", "wb"))
pickle.dump(vectorizer, open("vectorizer.pkl", "wb"))

print("✅ Model & Vectorizer saved!")
```

```python
# STEP 1: Import libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import pickle

# STEP 2: Load cleaned dataset
df = pd.read_csv("processed_nlp_data.csv")

# STEP 3: Select features
X = df["final_text"]       # Input text
y = df["final_subject"]    # Labels (category)

# STEP 4: Vectorization (IMPROVED)
vectorizer = TfidfVectorizer(
    ngram_range=(1,2),
    stop_words='english',
    max_features=5000
)

X_vec = vectorizer.fit_transform(X)

# STEP 5: Train model (IMPROVED)
model = LogisticRegression(max_iter=500)
model.fit(X_vec, y)

# STEP 6: Save model
pickle.dump(model, open("model.pkl", "wb"))
pickle.dump(vectorizer, open("vectorizer.pkl", "wb"))

print("✅ Model trained on REAL dataset successfully!")
```

```python
def map_category(label):
  label = str(label).lower()

  if"refund" in label:
    return "refund"
  elif "billing" in label or "payment" in label:
    return "billing"
  elif "delivery" in label or "shipping" in label:
    return "delivery"
  elif "product" in label:
    return "product"
  else:
    return "technical"
```

```python
df["category"] = df["final_subject"].apply(map_category)
```

```python
df[["final_subject" , "category"]].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import pickle
# features and labels
X = df["final_text"]
y = df["category"]
# split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# vectorizer
vectorizer = TfidfVectorizer()
x_train_vec = vectorizer.fit_transform(x_train)
# model
model = LogisticRegression(max_iter=300)
model.fit(x_train_vec, y_train)
# save
pickle.dump(model, open("model.pkl", "wb"))
pickle.dump(vectorizer, open("vectorizer.pkl", "wb"))
print("✅ Model trained successfully!")
```

```python
import pickle

# load model and vectorizer
model = pickle.load(open("model.pkl", "rb"))
vectorizer = pickle.load(open("vectorizer.pkl", "rb"))

# test input
text = ["My internet is not working properly"]

# transform
text_vec = vectorizer.transform(text)

# prediction
prediction = model.predict(text_vec)

print("Predicted Category:", prediction[0])
```
