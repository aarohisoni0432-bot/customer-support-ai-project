```python
# 📌 Step 1: Load Clean Dataset
# Loading cleaned data for NLP preprocessing

import pandas as pd

df = pd.read_csv('/content/cleaned_data_day1.csv')

df.head()
```

```python
# 📌 Step 2: Basic Text Cleaning
# This step cleans the text data by:
# 1. Converting text to lowercase
# 2. Removing special characters, numbers, and punctuation
# This helps standardize text for NLP models

import re

def clean_text(text):
    text = text.lower()  # convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # remove special characters
    return text

# Apply cleaning on both text columns
df['cleaned_description'] = df['Ticket Description'].apply(clean_text)
df['cleaned_subject'] = df['Ticket Subject'].apply(clean_text)

# Check output
df[['Ticket Description', 'cleaned_description']].head()
```

```python
# 📌 Step 3: Stopwords Removal
# This step removes common words (stopwords) like "the", "is", "and"
# which do not contribute much meaning to the text.
# Removing them helps improve model performance.

import nltk
from nltk.corpus import stopwords

# Download stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in stop_words])

# Apply stopword removal
df['final_description'] = df['cleaned_description'].apply(remove_stopwords)
df['final_subject'] = df['cleaned_subject'].apply(remove_stopwords)

# Check output
df[['cleaned_description', 'final_description']].head()
```

```python
# 📌 Step 4: Lemmatization
# This step converts words into their base/root form.
# It helps normalize text and improves model understanding.

import nltk
from nltk.stem import WordNetLemmatizer

# Download resources
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

# Apply lemmatization
df['lemmatized_description'] = df['final_description'].apply(lemmatize_text)
df['lemmatized_subject'] = df['final_subject'].apply(lemmatize_text)

# Check output
df[['final_description', 'lemmatized_description']].head()
```

```python
# 📌 Step 5: Combine Subject and Description
# This step combines the subject and description into a single text column.
# This gives the model more context for better predictions.

df['final_text'] = df['lemmatized_subject'] + " " + df['lemmatized_description']

# Check output
df[['lemmatized_subject', 'lemmatized_description', 'final_text']].head()
```

```python
# 📌 Step 6: Save Final NLP Dataset
# This step saves the fully processed text data.
# This dataset will be used for model training (ML + Deep Learning + BERT).

df.to_csv('processed_nlp_data.csv', index=False)
```
